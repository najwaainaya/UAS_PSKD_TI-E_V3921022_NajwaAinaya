```python
# ENKRIPSI DAN DESKRIPSI FILE CSV

# melakukan import modul
from cryptography.fernet import Fernet


# sebagai kunci
key = Fernet.generate_key()
 
# digunakan untuk membaca string kunci
with open('filekey.key', 'wb') as filekey:
   filekey.write(key)

# menggunakan kunci
fernet = Fernet(key)
 
# digunakan untuk membuka file
with open('C:/Users/User/Downloads/file_csv.csv', 'rb') as file:
    original = file.read()
     
# digunakan untuk mengenkripsi 
encrypted = fernet.encrypt(original)
 
# digunakan untuk menyimpan file enkripsi
with open('C:/Users/User/Downloads/file_csvenkripsi.csv', 'wb') as encrypted_file:
    encrypted_file.write(encrypted)


#DEKRIPSI
    
# menggunakan kunci yang sebelumnya sudah diinputkan
fernet = Fernet(key)

# digunakan untuk membuka file hasil enkripsi
with open('C:/Users/User/Downloads/file_csvenkripsi.csv', 'rb') as enc_file:
	encrypted = enc_file.read()

# digunakan untuk mendeskripsikan
decrypted = fernet.decrypt(encrypted)

# digunakan untuk mengetahui hasil setelah di deskripsi
with open('C:/Users/User/Downloads/file_csvdeskripsi.csv', 'wb') as dec_file:
	dec_file.write(decrypted)
```

```python
# ENKRIPSI PDF

from PyPDF2 import PdfFileWriter, PdfFileReader
  
# digunakan untuk membuat file
out = PdfFileWriter()
  
# digunakan untuk membuka file pdf yang asli
file = PdfFileReader("C:/Users/User/Downloads/file_pdf.pdf")
  
# digunakan untuk mengidentifikasi total halaman file
num = file.numPages
  
#digunakan untuk membaca setiap halaman file sesuai halaman yg diidentifikasi 
for idx in range(num):
    
   
    page = file.getPage(idx)
      

    out.addPage(page)
  
  
# digunakan untuk memasukkan password enkripsi 
password = "najwa"
  
# digunakan untuk enkripsi setiap halaman 
out.encrypt(password)
  
# membuka file enkripsi 
with open("C:/Users/User/Downloads/file_pdf_enkripsi.pdf", "wb") as f:
    
    # digunakan untuk menyimpan pdf  
    out.write(f)
```

```python
# DESKRIPSI PDF

from PyPDF2 import PdfFileWriter, PdfFileReader
  
# digunakan untuk membuat file
out = PdfFileWriter()
  
# digunakan untuk membuka file pdf yg terenkripsi
file = PdfFileReader("C:/Users/User/Downloads/file_pdf_enkripsi.pdf")
  
# digunakan untuk memasukkan password enkripsi 
password = "najwa"
  
# digunakan untuk cek file terenkripsi atau tidak 
if file.isEncrypted:
  
    # apabila file terenkripsi, dapat langsung di dekripsi menggunakan password 
    file.decrypt(password)
  
    # menggunakan dekripsi yang dilakukan setiap halaman file pdf
    # digunakan untuk menyimpan ke dalam file baru 
    for idx in range(file.numPages):
        
        # digunakan untuk mengidentifikasi halaman file 
        page = file.getPage(idx)
          
        # digunakan untuk memasukkan halaman yg sudah diidentifikasi dan sudah di dekripsi ke file baru 
        out.addPage(page)
      
    # membuka file baru 
    with open("C:/Users/User/Downloads/file_pdf_deskripsi.pdf", "wb") as f:
        
        # digunakan menyimpan file baru 
        out.write(f)
  

    print("File decrypted Successfully.")
else:
    
    print("File already decrypted.")
```

```python
try:
#    ENKRIPSI IMAGE

    # digunakan untuk menginputkan letak file image
    path = input(r'Enter path of Image : ') 
    # digunakan untuk menginputkan key dari image
    key = int(input('Enter Key for encryption of Image : '))
    
    # digunakan untuk memunculkan output dari letak file image
    print('The path of file : ', path)
    # digunakan untuk memunculkan output key dari image
    print('Key for encryption : ', key) 
    
    # membuka file path dengan mode biner
    fin = open(path, 'rb')
     
    image = fin.read()
    fin.close()
    
    #digunakan untuk mengembalikan mutable berupa rangkaian baris byte (byte array)  dari suatu objek  
    image = bytearray(image)
    
    # digunakan untuk membuat enkripsi dari image
    for index, values in enumerate(image):
        image[index] = values ^ key
 
    fin = open(path, 'wb')
     
    fin.write(image)
    fin.close()
    print('Encryption Done...')
  
except Exception:
    print('Error caught : ', Exception.__name__)
```

```python

try:
# DEKRIPSI IMAGE
    #digunakan untuk menginputkan letak file image
    path = input(r'Enter path of Image : ')
    # digunakan untuk menginputkan key dari image 
    key = int(input('Enter Key for encryption of Image : '))
        
    print('The path of file : ', path)
    print('Note : Encryption key and Decryption key must be same.')
    print('Key for Decryption : ', key)
     
    fin = open(path, 'rb')
     
    image = fin.read()
    fin.close()
        
    image = bytearray(image)
 
    for index, values in enumerate(image):
        image[index] = values ^ key
     
    fin = open(path, 'wb')
     
    fin.write(image)
    fin.close()
    print('Decryption Done...')
 
 
except Exception:
    print('Error caught : ', Exception.__name__)
```
